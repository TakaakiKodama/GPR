```julia
f(x)=2*x[1]
n(x)=[0.1]
N=10
X,Y=GPLM.ToyData(f,1,1;min=0,max=1,N=N,messiness=n)
plot(X,Y,st=:scatter)
```

```julia
k(x,x_,)=GPLM.GaussianKernel(x,x_,;th1=1,th2=1)
K=GPLM.KernelMatrix(X,k,observationn_noise=0.01)
# heatmap(K)
xs = [string("x", i) for i = 1:N]
ys = [string("y", i) for i = 1:N]
heatmap(xs, ys, K, aspect_ratio=1)
```

```julia
(μ,σ,K,K_,yy)=GPLM.GaussianProcessRegression(X,Y,k,x_test=[i for i in 0:0.01:1],observationn_noise=0.0001)
```

```julia
x_test=[i for i in 0:0.01:1]
x=[i for i in 0:0.1:1]
y=[f(i) for i in x]
plot(x,y)
plot!(x_test,μ,label="predicr")
plot!(X,Y,seriestype=:scatter)
plot!(x_test,μ+2 .*σ)
plot!(x_test,μ-2 .*σ)
```
